```python
import tensorflow as tf
from tensorflow import _keras
from keras.layers import Dense,Flatten
from keras.models import Sequential
from sklearn.model_selection import train_test_split
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv('6_gradient_descent\insurance_data.csv')
df.head()
```

```python
df.age = df.age/100
```

```python
X_train,X_test , y_train,y_test = train_test_split(df[['age','affordibility']] , df.bought_insurance , test_size=0.25)
```

```python
model = Sequential([
    Flatten(input_shape = (2,)),
    Dense(1)
])
```

```python
model.compile(optimizer='adam' , loss = 'binary_crossentropy' , metrics=['accuracy'])
```

```python
model.fit(X_train,y_train , epochs=500)
```

```python
model.evaluate(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred_label = [np.argmax(i) for i in y_pred]
```

```python

from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred_label))
```
